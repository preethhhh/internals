```python
import pandas as pd
a=pd.Series([10,20,30,40,50])
b=pd.Series([40,50,60,70,80])
print("series a:",a)
print("series b:",b)
non_com = a[~a.isin(b)].tolist() + b[~b.isin(a)].tolist()
print("non common ",non_com)
print("Smallest  in  A:", a.min())
print("largest  in  A:", a.max())
print("Sum of  B:", b.sum())
print("Average of  A:",a.mean())
print("Median of B:", b.median())
```

```python
#box plot
import seaborn as s
import matplotlib.pyplot as plt
import pandas as pd
d= s.load_dataset('titanic')
s.boxplot(x=d['survived'], y=d['age'])
plt.show()
```

```python
#pie chart
import pandas as pd
import matplotlib.pyplot as plt
p = pd.read_csv("titanic.csv")
d = pd.DataFrame(p)
survived_count = d['Survived'].sum()
not_survived_count = len(d) - survived_count
labels = ['Survived', 'Not Survived']
sizes = [survived_count, not_survived_count]
colors = ['b','r']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Survival Rate of Titanic Passengers')
```

```python
#bar chart
import pandas as p
import matplotlib.pyplot as m
d=p.read_csv("tested.csv")
c=d["Pclass"].value_counts()
co=['g','r','b']
m.bar(c.index,c.values,color=co,width=0.5)
m.xticks([1,2,3],["1st class","2nd class","3rd class"])
m.xlabel("Classes");m.ylabel("No of Passengers");m.title("No of Passengers Travelled in Specific Classes")
m.show()
```

```python
#scatter plot
import pandas as pd
import matplotlib.pyplot as plt
p = pd.read_csv("titanic.csv")
d = pd.DataFrame(p)
plt.scatter(d['Age'], d['Fare'], alpha=0.5, c=d['Survived'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs. Fare')
plt.show()
```
